# Part 1 - Data Cleaning

## Data Loading
We load the data, provided in `hw2-crowd-data.csv`

```python
df_cars = pd.read_csv("hw2-crowd-data.csv")
```

```python
# what is the length of my dataframe?
len(df_cars)
```

## Data Cleaning
You need to clean the rows and columns of data using Pandas.  Insert the code (and descriptions of the code in markdown) in this subsection.

```python
#Drop missing valued rows
df_cars = df_cars.dropna()
```

```python
#Drop duplicate rows
df_cars = df_cars.drop_duplicates()
```

##### At this point, we have cleaned the entire dataset of rows with missing values or duplicate rows. We now need to fix the formatting of the columns, but we recognize that we only need the columns of Price and Mileage to answer our research question. Therefore, we focus on reformatting those columns in specific, by converting all of the values in those columns to integers.

```python
#Fix poorly formatted price values by making all values integers
df_cars['Price'] = df_cars['Price'].astype(int)

#Fix poorly formatted mileage values by making all values integers
df_cars['Mileage'] =  df_cars['Mileage'].astype(int)
```

```python
#Fix poorly formatted string values 

for col in ['Car Make', 'Car Model', 'Dealer or Individual', 'Location', 'Color', 'Transmission', 'VIN#', 'Engine', 'Description', 'Web link', 'Screenshot']:
    df_cars[col] =  df_cars[col].astype(str)

#Fix poorly formatted door values by making all integers (if potentially string)
df_cars['Doors'] = df_cars['Doors'].apply(lambda x: 4 if isinstance(x,int) or (isinstance(x,str)==True and x and x[0] == "4") else "")

#Fix poorly formatted year values by making all values integers
for col in ['Year']:
    df_cars[col] =  df_cars[col].astype(int)
```

## Export Cleaned Data
Save the output of your data cleaning to `hw2-clean-data.csv`

```python
df_cars.to_csv("hw2-clean-data.csv")
```
